```python
"""
Write a Python Program to implement your own myreduce() function which works exactly
like Python's built-in function reduce()
"""

def my_reduce(anyfunc, i_list):

 # Get first item in sequence and assign to result
  result = i_list[0]
 # iterate over remaining items in sequence and apply reduction function 
  for num in i_list[1:]:
   result = anyfunc(result, num)
  return result

# test myreduce function
def mul(x,y): 
    return x * y

input_list=[1,2,3,4,5]

print ("Multiplication using custom reduce function is : " + str(my_reduce(mul,input_list)) )
```

```python
'''Write a Python program to implement your own myfilter() function which works exactly
like Python's built-in function filter()'''


def my_filter(anyfunc, ilist):

 result = []  
 for num in ilist:      # iterate over sequence of items in sequence and apply filter function
  if anyfunc(num):
   result.append(num)
 return result

# test myfilter function
def check_even(x):
    return x%2==0

input_list=[-6,1,5,-3,-8,9,7,4,8]

print ("Even numbers using custom filter function are :"  + str(my_filter(check_even, input_list)))
```

```python
'''Implement a function longestWord() that takes a list of words and returns the longest one. '''

def longestWord(*args):
    long_word=args[0]
    for i in range(len(args)):
        if len(args[i]) > len(long_word):
            long_word=args[i]
    return "longest word is " + long_word +" with length " +str(len(long_word))
        
```

```python
longestWord('ab','cdfg','rgd','a','aaaaa')
```

```python
'''Write a function filter_long_words() that takes a list of words 
and an integer n and returns the list of words that are longer than n. '''

def filter_long_words(*args):
    for i in range(len(args)):
        if type(args[i])==int:
            n=args[i]
    l=[]
    for i in range(len(args)):
        if type(args[i])==str:
            if len(args[i]) >n:
                l.append(args[i])
    return l
    
```

```python
filter_long_words('aaa',4,'bbbbb','cccccc','asd','aaaaaaa','fgh')
```
